```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score, roc_auc_score
```

```python
# dataset repository path
url = "https://raw.githubusercontent.com/gaeng02/KHUDA-9th-ML/main/week4/dataset.csv"
```

```python
df = pd.read_csv(url)
df.head()
```

```python
### answer init
answer = [0] * 5
```

## Task 0. 초기 설정
다음 설정에 맞게 모델을 만드세요. <br>
별도의 정답은 없지만, 완료되지 않은 경우 Task 수행 불가 <br>
`X_train, y_train`을 이용하여 학습까지 진행하세요.

- feature(X) : `x1~x8`
- target(y) : `target`
- split : `test_size = 0.2, random_state = 212, stratify = y`
- model : `LogisticRegression(max_iter =2000, solver = "lbfgs")`

<br>

이때, X는 `StandardScaler`를 이용하여 데이터 전처리를 진행해주세요.

```python
### Step 1. feature, target Setting
feature_cols = [f"x{i}" for i in range (1, 9)]
X =df[feature_cols]
y =df['target']

### Step 2. Data Split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=212,stratify=y)


### Step 3. Data Scaling
ss=StandardScaler()
xtrain_scaled=ss.fit_transform(X_train)
xtest_scaled=ss.transform(X_test)

### Step 4. Modeling
lg=LogisticRegression(max_iter=2000,solver="lbfgs")
lg.fit(xtrain_scaled,y_train)

```

## Task 1
test dataset(`X_test`)을 이용하여 Accuracy를 구하세요.  
이때 Accuracy를 `answer[1]` 저장하세요.

```python
## Input Box

accuracy = 0.0
pred=lg.predict(xtest_scaled)
accuracy=accuracy_score(y_test,pred)



answer[1] = accuracy
print(answer[1])
```
